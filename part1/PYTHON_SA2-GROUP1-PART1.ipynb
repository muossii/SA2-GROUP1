```python
#Rename month names to English.
month_map = {'Janeiro':'January', 'Feveriro':'February', 
             'Marco': 'March', 'Abril':'April', 'Maio':'May', 
             'Junho': 'June', 'Julho': 'July', 'Agosto':'August',
             'Setembro': 'September', 'Outubro': 'October', 
             'Novembro': 'November', 'Dezembro': 'December'}
df['month'] = df['month'].map(month_map)
```

```python
df.number.sum()
```

The total number of fires register is 698811.073.

```python
#In which month maximum number of forest fires were reported?
pivot1 = pd.pivot_table(df,values='number',index=['month'],aggfunc=np.sum)

pivot1
```

```python
#plotting the table with a bar graph

mpl.figure(figsize=(15, 6))
sns.set_style('darkgrid')
ax = sns.barplot(x=pivot1.index, y= 'number', data=pivot1)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
mpl.xlabel('Month')
mpl.ylabel('Number of fires')
mpl.title('Fire vs Month')
```

From the graph, July has the reported maximum number of forest fires.

```python
#In which year maximum number of forest fires was reported?

pivot2 = pd.pivot_table(df,values='number',index=['year'],aggfunc=np.sum)

pivot2
```

```python
#plotting the table with a bar graph

mpl.figure(figsize=(15, 6))
sns.set_style('darkgrid')
ax = sns.barplot(x=pivot2.index, y= 'number', data=pivot2, color = 'coral')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
mpl.xlabel('Year')
mpl.ylabel('Number of fires')
mpl.title('Number of Forest Fires by Year')
```

From the data, forest fires have been increasing from 1998, and it had a rapid increase in 2002. It reached its maximum peak at 2003, and lost its momentum right after 5 years. However, forest fires increased again in 2009, going up and down from the range of 35,000 to 40,000 fires up until 2017.

```python
#In which state maximum number of forest fires was reported?

pivot3 = pd.pivot_table(df,values='number',index=['state'],aggfunc=np.sum)

pivot3
```

```python
#Plotting the graph
mpl.figure(figsize=(15, 6))
sns.set_style('darkgrid')
ax = sns.barplot(x=pivot3.index, y= 'number', data=pivot3)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
mpl.xlabel('State')
mpl.ylabel('Number of fires')
mpl.title('Number of Forest Fires by State')
```

Mato Grosso has the most number of forest fires, with a record number of 96,246.028.

```python
#Find total number of fires were reported in Amazonas.
amazonas_df = df[df['state'] == 'Amazonas']
total_fires = amazonas_df['number'].sum()
print('The total number of fires reported in Amazonas is', total_fires)
```

```python
#Display number of fires were reported in Amazonas (year-wise).
#plotting a graph
mpl.figure(figsize=(15, 6))
sns.set_style('darkgrid')
mpl.bar(amazonas_df['year'], amazonas_df['number'], color = 'coral')
mpl.title('Number of Fires Reported in Amazonas')
mpl.xlabel('Year')
mpl.ylabel('Number of Fires')
mpl.show()
```

Based from the graph, the number of fires reported in Amazonas has been increasing over the years with the maximum number of fires being reported in 2007. It also seems there is a cycle pattern in the number of fires being reported, with a peak occuring in a year followed by a decline. Overall, the graph suggests that forest fires have been a persistent problem in Amazonas, with occasional peaks in the number of fires being reported. 

```python
#Display number of fires were reported in Amazonas (day-wise).
amazonas = df[df['state']== 'Amazonas']
date_amazonas = amazonas.groupby('date')['number'].sum()

date_amazonas
```

```python
#Find total number of fires were reported in 2015 and visualize data based on each â€˜monthâ€™.
df_2015 = df[df['year'] == 2015]
fires_month = df_2015.groupby('month')['number'].sum()
mpl.figure(figsize=(15, 6))
sns.set_style('darkgrid')
mpl.bar(fires_month.index, fires_month.values, color = 'coral')
mpl.title('Number of Fires Reported in Amazon in 2015 by Month')
mpl.xlabel('Month')
mpl.ylabel('Number of Fires')
mpl.show()
```

```python
#Find average number of fires were reported from highest to lowest (state-wise).
state_ave = df.groupby('state')['number'].mean().reset_index()

state_ave = state_ave.sort_values(by='number', ascending=False)

state_ave
```
